```python
import pandas as pd

data = pd.read_csv("iris.csv")
print(data.head())
```

```python
print("Dataset information")
display(data.info())
```

```python
print("Missing values")
display(data.isnull().sum())
```

```python
print("Basic statistics for numerical columns:")
display(data.describe())
```

```python
print("Grouping by species")
display(data.groupby('species').mean())
```

```python
#Interesting findings
data['sepal_area'] = data ['sepal_length'] * data['sepal_width']
largest_sepal = data.loc[data['sepal_area'].idxmax()]
print(f"Flower with largest sepal area: {largest_sepal['species']} (ID: {largest_sepal.name})")
```

```python
import matplotlib.pyplot as plt

graph = data.groupby('species')['petal_length'].mean()

graph.plot.bar(color = ['blue', 'green', 'pink'])
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal length')

```

```python
import seaborn as sns 

sns.histplot(data=data, x='sepal_width', bins=15, kde=True, hue='species', palette='viridis')
plt.title('Distribution of Sepal Width by Species')
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
```

```python
sns.scatterplot(data=data, x='sepal_length', y='petal_length', hue='species', palette='viridis', s=100)
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend(title='Species')
```
